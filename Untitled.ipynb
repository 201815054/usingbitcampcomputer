## 01) NLP에서의 사전 훈련(Pre -training)
- 사전 훈련된 단어 임베딩의 모든 NLP 실무자의 도구 상자에서 
    사전 훈련된 언어 모델로 대체되는것을 시간문제이다.

### 1. 사전 훈련된 워드 임베딩
- 워드 임베딩 방법론 : Word2Vec, FastText, GloVe
- 임베딩을 사용하는 방법에는 크게 두가지가있음 
- => Embedding layer을 랜덤 초기화하여 처음부터 학습/ 사전에 학습된 임베딩 벡터들을 가져와 사용 (데이터가 적을때 성능향상)
- But 문제점이 있음. 하나의 단어가 하나의 벡터값으로 맵핑이 되어서 다의어나 동음이의어를 구분하지 못함. 
- => ex) 사과엔 두가지 의미가있지만 하나의 벡터값을 맵핑하므로 두가지의미를 구분할 수 없음. 
- 이 문제점은 사전 훈련된 언어 모델을 사용함으로써 극복 할 수 있음.
